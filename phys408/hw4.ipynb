```python
import sympy as sp
from IPython.display import display, Math

# Define symbolic variables
r, theta, phi = sp.symbols("r theta phi", real=True, positive=True)
q, epsilon0 = sp.symbols("q epsilon_0", real=True, positive=True)
px, py, pz = sp.symbols("p_x p_y p_z", real=True)
Qxx, Qyy, Qzz = sp.symbols("Q_xx Q_yy Q_zz", real=True)

# Define unit vectors in spherical coordinates
x = r * sp.sin(theta) * sp.cos(phi)
y = r * sp.sin(theta) * sp.sin(phi)
z = r * sp.cos(theta)

# Define the potential up to quadrupole term
monopole = q / r
dipole = (px * x + py * y + pz * z) / r**3
quadrupole = (
    (1 / 2)
    * (Qxx * (3 * x**2 - r**2) + Qyy * (3 * y**2 - r**2) + Qzz * (3 * z**2 - r**2))
    / r**5
)

potential = (1 / (4 * sp.pi * epsilon0)) * (monopole + dipole + quadrupole)

# Calculate electric field components in spherical coordinates
Er = -sp.diff(potential, r)
Etheta = -(1 / r) * sp.diff(potential, theta)
Ephi = -(1 / (r * sp.sin(theta))) * sp.diff(potential, phi)

# Simplify expressions
Er = sp.simplify(Er)
Etheta = sp.simplify(Etheta)
Ephi = sp.simplify(Ephi)

# Print results in LaTeX format
print("Potential:")
display(Math(sp.latex(potential)))
print("\nElectric Field (r-component):")
display(Math(sp.latex(Er)))
print("\nElectric Field (theta-component):")
display(Math(sp.latex(Etheta)))
print("\nElectric Field (phi-component):")
display(Math(sp.latex(Ephi)))
```

```python
# Set up symbolic matrix for inversion
# Define symbols for quadrupole moment matrix
Qxx, Qyy, Qzz = sp.symbols("Q_xx Q_yy Q_zz", real=True)
A = sp.Matrix([[Qxx, 0, 0], [0, Qyy, 0], [0, 0, Qzz]])

# Compute the inverse of the matrix symbolically
A_inv = A.inv()

# Display the inverse matrix
print("Inverse of the quadrupole moment matrix:")
display(Math(sp.latex(A_inv)))
```
