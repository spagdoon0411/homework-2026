```python
# Define symbolic variables
alpha, beta, gamma, E = sympy.symbols("alpha beta gamma E")

# Define the vector field E
E_field = sympy.Matrix([0, 0, E])

# Define the path of integration from (0,0,0) to (alpha, beta, gamma)
start_point = sympy.Matrix([0, 0, 0])
end_point = sympy.Matrix([alpha, beta, gamma])

# Perform line integral by integrating along the path
# Integrate -E along the z-axis from 0 to E
line_integral = sympy.integrate(-E, (E, 0, end_point[2]))

# Simplify the result
result = sympy.simplify(line_integral)

print("Line integral of -E from (0,0,0) to (α,β,γ):", result)
```

```python
# Define symbolic variables for the problem
x, y, a, c, t, theta = sympy.symbols("x y a c t theta")

# Define the force vector F
F_x = c * y
F_y = -c * x
F_vector = sympy.Matrix([F_x, F_y])


# Path C1: Two-segment path
# First segment: x from 0 to a, y = 0
# Second segment: y from 0 to a, x = a
def calculate_work_C1():
    # First segment integral along x-axis
    work_x = sympy.integrate(F_x.subs(y, 0), (x, 0, a))
    print("work_x", work_x)

    # Second segment integral along y-axis
    work_y = sympy.integrate(F_y.subs(x, a), (y, 0, a))
    print("work_y", work_y)

    return sympy.simplify(work_x + work_y)


# Path C2: Direct diagonal line from (0,0) to (a,a)
def calculate_work_C2():
    # Parametric representation of the line
    x_param = t * a
    y_param = t * a

    # Substitute parametric equations into F_x and F_y
    F_x_param = c * (t * a)
    F_y_param = -c * (t * a)

    # Integrate along the parameter t from 0 to 1
    work = sympy.integrate(
        F_x_param * sympy.diff(x_param, t) + F_y_param * sympy.diff(y_param, t),
        (t, 0, 1),
    )

    return sympy.simplify(work)


# Path C3: Parametric curve
def calculate_work_C3():
    # Parametric equations
    x_param = a * (1 - sympy.cos(theta))
    y_param = a * sympy.sin(theta)

    # Derivatives
    dx_dtheta = a * sympy.sin(theta)
    dy_dtheta = a * sympy.cos(theta)

    # Substitute parametric equations into F_x and F_y
    F_x_param = c * (a * sympy.sin(theta))
    F_y_param = -c * (a * (1 - sympy.cos(theta)))

    # Integrate along theta from 0 to pi/2
    work = sympy.integrate(
        F_x_param * dx_dtheta + F_y_param * dy_dtheta, (theta, 0, sympy.pi / 2)
    )

    return sympy.simplify(work)


# Calculate work for each path
work_C1 = calculate_work_C1()
work_C2 = calculate_work_C2()
work_C3 = calculate_work_C3()

print("Work along path C1:", work_C1)
print("Work along path C2:", work_C2)
print("Work along path C3:", work_C3)

# Attempt to find potential energy U(x,y)
# If F is conservative, curl of F should be zero
curl_F = sympy.diff(F_y, x) - sympy.diff(F_x, y)
print("\nCurl of F:", curl_F)


# If curl is zero, we can find potential U by integrating
def find_potential():
    U = sympy.integrate(F_x, x) + sympy.Symbol("C")
    return U


potential_U = find_potential()
print("\nPotential Energy U(x,y):", potential_U)
```

```python
# Define symbolic variables
r, theta, k, R = sympy.symbols("r theta k R")

# Define V(r, theta)
V = -k / 3 + (2 * k / (3 * R**2)) * r**2 * (3 * sympy.cos(theta) ** 2 - 1)

# Define W(r, theta)
W = -k * R / (3 * r) + (2 * k * R**3) / (3 * r**3) * (3 * sympy.cos(theta) ** 2 - 1)

# Calculate gradient of V
dV_dr = sympy.diff(V, r)
dV_dtheta = sympy.diff(V, theta)

# Calculate gradient of W
dW_dr = sympy.diff(W, r)
dW_dtheta = sympy.diff(W, theta)

# Display gradients in LaTeX
from IPython.display import display, Math

print("Gradient of V(r, θ):")
display(Math(r"\frac{\partial V}{\partial r} = " + sympy.latex(dV_dr)))
display(Math(r"\frac{\partial V}{\partial \theta} = " + sympy.latex(dV_dtheta)))

print("\nGradient of W(r, θ):")
display(Math(r"\frac{\partial W}{\partial r} = " + sympy.latex(dW_dr)))
display(Math(r"\frac{\partial W}{\partial \theta} = " + sympy.latex(dW_dtheta)))
```

```python
# Substitute r = R before simplifying
dW_dr_at_R = dW_dr.subs(r, R)
dV_dr_at_R = dV_dr.subs(r, R)

display(
    Math(
        r"\frac{\partial W}{\partial r} - \frac{\partial V}{\partial r} \Big|_{r=R} = "
        + sympy.latex(sympy.simplify(dW_dr_at_R - dV_dr_at_R))
    )
)
```

```python
# Define piecewise potential function V(r, θ)
k, r, R, theta = sympy.symbols('k r R theta')

V_inner = sympy.Rational(-3, 5) * k * (r/R) * sympy.cos(theta) + sympy.Rational(4, 5) * k * (r/R)**3 * (5 * sympy.cos(theta)**3 - 3 * sympy.cos(theta))
V_outer = sympy.Rational(-3, 5) * k * (R/r)**2 * sympy.cos(theta) + sympy.Rational(4, 5) * k * (R/r)**4 * (5 * sympy.cos(theta)**3 - 3 * sympy.cos(theta))

V = sympy.Piecewise((V_inner, r <= R), (V_outer, r > R))

# Calculate gradients
dV_dr = sympy.diff(V, r)
dV_dtheta = sympy.diff(V, theta)

# Display potential and gradients
print("Potential V(r, θ):")
display(Math(r"V(r, \theta) = " + sympy.latex(V)))

print("\nGradient of V(r, θ):")
display(Math(r"\frac{\partial V}{\partial r} = " + sympy.latex(dV_dr)))
display(Math(r"\frac{\partial V}{\partial \theta} = " + sympy.latex(dV_dtheta)))
```

```python
# Compute the difference in dV/dr at r = R
dV_dr_inner_at_R = dV_dr.subs(r, R).subs([(r, R), (theta, sympy.Symbol('theta'))])
dV_dr_outer_at_R = dV_dr.subs(r, R).subs([(r, R), (theta, sympy.Symbol('theta'))])
dV_dr_diff_at_R = dV_dr_inner_at_R - dV_dr_outer_at_R

print("\nDifference in dV/dr at r = R:")
display(Math(r"\left.\frac{\partial V}{\partial r}\right|_{r=R^-} - \left.\frac{\partial V}{\partial r}\right|_{r=R^+} = " + sympy.latex(dV_dr_diff_at_R)))

```
