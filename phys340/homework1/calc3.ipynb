```python
import math


def round_uncertainty_numeric(value, uncertainty, cutoff=2):
    exp = math.floor(math.log10(abs(uncertainty)))
    first_digit = int(uncertainty / 10**exp)
    sigs = 2 if first_digit <= cutoff else 1
    rounded_uncertainty = round(
        uncertainty, -int(math.floor(math.log10(uncertainty))) + (sigs - 1)
    )
    decimals = -int(math.floor(math.log10(rounded_uncertainty))) + (sigs - 1)
    rounded_value = round(value, decimals)
    return rounded_value, rounded_uncertainty


v, u = round_uncertainty_numeric(12.345, 0.1267)
print(v, u)
```

```python
import numpy as np


def round_unc(value, uncertainty, cutoff=2):
    scalar_input = np.isscalar(value)

    value = np.atleast_1d(value).astype(float)
    uncertainty = np.atleast_1d(uncertainty).astype(float)

    exp = np.floor(np.log10(np.abs(uncertainty)))
    first_digit = (uncertainty / 10**exp).astype(int)
    sigs = np.where(first_digit <= cutoff, 2, 1)

    rounding_decimals = (-exp + (sigs - 1)).astype(int)
    rounded_uncertainty = (
        np.round(uncertainty * 10.0**rounding_decimals) / 10.0**rounding_decimals
    )

    decimals = (-np.floor(np.log10(rounded_uncertainty)) + (sigs - 1)).astype(int)
    rounded_value = np.round(value * 10.0**decimals) / 10.0**decimals

    if scalar_input:
        return rounded_value[0], rounded_uncertainty[0]
    return rounded_value, rounded_uncertainty
```

```python
values = np.array([12.3432, 12.345])
uncertainties = np.array([0.31, 0.1267])
round_unc(values, uncertainties)
```

```python
import numpy as np

# Example: Large value with large uncertainty
value = np.array([1234.5])
uncertainty = np.array([56.7])

# Following the algorithm:
# exp = 1, first_digit = 5, sigs = 1 (since 5 > cutoff=2)
# rounded_uncertainty = 60
# decimals = -floor(log10(60)) + 0 = -1

decimals = np.array([-1])

# This FAILS with integer 10:
try:
    result = value * 10**decimals
except ValueError as e:
    print(f"Error with 10**decimals: {e}")

# This WORKS with float 10.0:
result = value * 10.0**decimals
print(f"Success with 10.0**decimals: {result}")  # [123.45]
```
