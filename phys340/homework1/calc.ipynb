```python
import pandas as pd
import numpy as np

df = pd.read_csv("2_12.csv")

df["disc"] = df["accel_value"] - df["expected_value"]
df["disc_unc"] = np.sqrt(df["accel_unc"] ** 2 + df["expected_unc"] ** 2)
df["discrep_over_unc"] = np.abs(df["disc"] / df["disc_unc"])

print(df)
```
