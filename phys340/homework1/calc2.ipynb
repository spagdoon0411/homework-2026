```python
import pandas as pd
import numpy as np

values = pd.read_csv("4_13.csv", header=None).values
df = pd.DataFrame()
df["values"] = values.flatten()
df.std(ddof=1)
```

```python
mean = df["values"].mean()
std = df["values"].std(ddof=1)
df["outside_1sigma"] = (df["values"] - mean).abs() >= std
df["outside_2sigma"] = (df["values"] - mean).abs() >= 2 * std
print("outside 1 sigma:", df["outside_1sigma"].sum())
print("outside 2 sigma:", df["outside_2sigma"].sum())
```

```python
expected_outside_1sigma = (1 - 0.68) * df["values"].count()
expected_outside_2sigma = (1 - 0.95) * df["values"].count()

print("expected outside 1 sigma:", expected_outside_1sigma)
print("expected outside 2 sigma:", expected_outside_2sigma)
```

```python
mean = df["values"].mean()
std = df["values"].std(ddof=1)
sdom = std / np.sqrt(df["values"].count())

print(mean, std, sdom)
```
