```python
import numpy as np

num = np.array(
    [
        4 * np.pi,
        8.854e-12,
        0.667e-30,
        500,
    ]
)

denom = np.array(
    [
        1e-3,
        1.602e-19,
    ]
)

num.prod() / denom.prod()
```
