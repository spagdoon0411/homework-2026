```python
import numpy as np


def main():
    # Given data
    voltages = np.array([11.2, 13.4, 15.1, 17.7])  # volts
    currents = np.array([4.67, 5.46, 6.28, 7.22])  # amps
    rated_resistance = 2.50  # ohms
    systematic_error_percent = 2.0  # %

    # Part (a): Calculate resistance values and random uncertainty
    resistances = voltages / currents
    best_r = np.mean(resistances)
    random_uncertainty = np.std(resistances, ddof=1) / np.sqrt(len(resistances))

    # Print part (a) results
    print("Part (a):")
    print(f"Individual resistance values (ohms): {resistances}")
    print(f"Best estimate for resistance: {best_r:.4f} ohms")
    print(f"Random uncertainty: {random_uncertainty:.4f} ohms")

    # Part (b): Error propagation for systematic uncertainty
    # For R = V/I, relative systematic uncertainty is sqrt((δV/V)^2 + (δI/I)^2)
    relative_sys_error = np.sqrt(2 * (systematic_error_percent / 100) ** 2)
    systematic_uncertainty = relative_sys_error * best_r

    # Combine uncertainties in quadrature
    total_uncertainty = np.sqrt(random_uncertainty**2 + systematic_uncertainty**2)

    # Final result with total uncertainty
    print("\nPart (b):")
    print(f"Systematic uncertainty: {systematic_uncertainty:.4f} ohms")
    print(f"Total uncertainty: {total_uncertainty:.4f} ohms")
    print(f"Final result: {best_r:.4f} ± {total_uncertainty:.4f} ohms")

    # Compare with rated value
    print("\nComparison with rated value:")
    print(f"Rated value: {rated_resistance} ohms")
    difference = abs(best_r - rated_resistance)
    print(f"Difference from rated value: {difference:.4f} ohms")

    if difference <= total_uncertainty:
        print("The rated value lies within the uncertainty range.")
    else:
        print("The rated value does not lie within the uncertainty range.")


if __name__ == "__main__":
    main()
```

```python
from uncertainties import ufloat
import numpy as np
import math

# Part (c): (2.0 ± 0.1) × tan(45 ± 3°)
A = ufloat(2.0, 0.1)
B_degrees = ufloat(45, 3)
B_radians = B_degrees * (np.pi / 180)  # Convert to radians for Python's tan function

# Step by step
D = np.tan(
    B_radians.nominal_value
)  # We need to handle tan separately due to degree-radian conversion
D_uncertainty = abs(1 / np.cos(B_radians.nominal_value) ** 2) * B_radians.std_dev
D_with_uncertainty = ufloat(D, D_uncertainty)
print(f"Tangent result: {D_with_uncertainty}")

result_c = A * D_with_uncertainty
print(f"Final result (c): {result_c}")  # Should be approximately 2.00 ± 0.23
```
