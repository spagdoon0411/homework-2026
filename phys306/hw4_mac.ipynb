```python
import numpy as np
import matplotlib.pyplot as plt

A = 1.0
m = 1.0
k = 1.0
k2 = 0.2
b = 0.04
beta = b / (2 * m)

omega1 = np.sqrt(k / m)
omega2 = np.sqrt((k + 2 * k2) / m)


def Aj(omegaj, beta):
    omegajp = np.sqrt(omegaj**2 - beta**2)
    return (A / 2) * omegaj / omegajp


def deltaj(omegaj, beta):
    return np.arctan(beta / np.sqrt(omegaj**2 - beta**2))


A1 = Aj(omega1, beta)
A2 = Aj(omega2, beta)

delta1 = deltaj(omega1, beta)
delta2 = deltaj(omega2, beta)


def solution(A1, A2, delta1, delta2, omega1, omega2, t):
    g1 = np.sqrt(omega1**2 - beta**2)
    g2 = np.sqrt(omega2**2 - beta**2)

    term1 = A1 * np.exp(-beta * t) * np.cos(g1 * t - delta1)
    term2 = A2 * np.exp(-beta * t) * np.cos(g2 * t - delta2)

    s1 = term1 + term2
    s2 = term1 - term2

    return np.array([s1, s2])


def solution_wrapped(t):
    return solution(A1, A2, delta1, delta2, omega1, omega2, t)


t = np.linspace(0, 80, 1000)
s = solution_wrapped(t)
s1, s2 = s

plt.figure(figsize=(12, 6))
plt.plot(t, s1, label="x1(t)")
plt.plot(t, s2, label="x2(t)")
plt.title("Coupled Oscillators with Damping")
plt.xlabel("Time (t)")
plt.ylabel("Displacement")
plt.legend()
plt.grid()
plt.show()
```
